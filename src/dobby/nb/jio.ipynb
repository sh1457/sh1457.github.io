```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = Path(globals()['_dh'][0]).parent.parent.parent / 'static' / 'data'
```

```python
def compare_scatter_plot(df: pd.DataFrame) -> None:
    def cmap(val: float) -> str:
        return {3.0: "red", 2.0: "yellow", 1.5: "green"}.get(val, "blue")
    df.assign(color=lambda x: x.data.apply(cmap)).plot(kind='scatter', x="price_per_day", y="data_per_day", s="validity", c="color")
    df.assign(color=lambda x: x.data.apply(cmap)).plot(kind='scatter', x="validity", y="price", s="price_per_day", c="color")
    df.assign(color=lambda x: x.data.apply(cmap)).plot(kind='scatter', x="validity", y="total_data", s="price_per_day", c="color")
```

```python
df = pd.read_csv(data_path / 'jio.csv')
compare_scatter_plot(df)
```
